# Spark Connection Example

This notebook demonstrates how to connect to the Spark cluster from Jupyter.


```python
# Import required libraries
from pyspark.sql import SparkSession

spark = SparkSession.builder.enableHiveSupport().getOrCreate()

print(f"Spark version: {spark.version}")
print(f"Spark master: {spark.sparkContext.master}")
print(f"Spark UI: http://localhost:8080")
print(f"Application UI: {spark.sparkContext.uiWebUrl}")
```

```python
# Test Spark connection with a simple operation
# Progress will be shown in the output below
print("Creating DataFrame...")
df = spark.createDataFrame([(1, "Alice"), (2, "Bob"), (3, "Charlie")], ["id", "name"])
print("DataFrame created successfully!")
df.show()
```
